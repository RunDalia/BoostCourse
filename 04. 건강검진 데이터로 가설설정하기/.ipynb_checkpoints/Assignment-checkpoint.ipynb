# . Ïó∞Î†πÎåÄÎ≥Ñ ÌóàÎ¶¨ÎëòÎ†àÏóê ÎåÄÌïú Í∏∞Ïà†ÌÜµÍ≥ÑÎ•º Íµ¨ÌïòÎ†§Í≥† Ìï©ÎãàÎã§. 
Îã§Ïùå Ï†úÍ≥µÎêòÎäî ÎîïÏÖîÎÑàÎ¶¨Î•º ÌÜµÌï¥ Ïó∞Î†πÎåÄÏΩîÎìú(5ÏÑ∏Îã®ÏúÑ)Î•º "Ïó∞Î†πÎåÄ"Î°ú ÎßåÎì§Í≥† ÏïÑÎûòÏôÄ Í∞ôÏùÄ Í∏∞Ïà†ÌÜµÍ≥ÑÍ∞íÏùÑ Íµ¨Ìï¥Ï£ºÏÑ∏Ïöî!

üîΩ ÎîïÏÖîÎÑàÎ¶¨ 

```python
age_code = {1: '0~4ÏÑ∏',
            2: '5~9ÏÑ∏',
            3: '10~14ÏÑ∏',
            4: '15~19ÏÑ∏',
            5: '20~24ÏÑ∏',
            6: '25~29ÏÑ∏',
            7: '30~34ÏÑ∏',
            8: '35~39ÏÑ∏',
            9: '40~44ÏÑ∏',
            10: '45~49ÏÑ∏',
            11: '50~54ÏÑ∏',
            12: '55~59ÏÑ∏',
            13: '60~64ÏÑ∏',
            14: '65~69ÏÑ∏',
            15: '70~74ÏÑ∏',
            16: '75~79ÏÑ∏',
            17: '80~84ÏÑ∏',
            18: '85ÏÑ∏+'}
```

üîΩ Ï∂úÎ†• ÏòàÏãú

Îã§Ïùå Ïù¥ÎØ∏ÏßÄÏôÄ Í∞ôÏùÄ Í≤∞Í≥ºÍ∞Ä ÎÇòÏò§Î©∞, ÏÜåÏà´Ï†ê ÌëúÌòÑ Î∞©ÏãùÏù¥ Îã§Î•¥ÎçîÎùºÎèÑ Ï∂úÎ†•Í∞íÏù¥ Í∞ôÏúºÎ©¥ ÏÉÅÍ¥Ä ÏóÜÏäµÎãàÎã§. ÏòàÎ•º Îì§Ïñ¥ ÏòàÏãúÍ≤∞Í≥ºÎäî 32Ïù∏Îç∞ ÏßÅÏ†ë Íµ¨Ìïú Í∞íÏù¥ 32.0 Î°ú Î≥¥Ïù¥ÎçîÎùºÎèÑ ÏÉÅÍ¥ÄÏù¥ ÏóÜÏäµÎãàÎã§.


```python
# mapÌï®ÏàòÎ°ú Ïó∞Î†πÎåÄ ÏΩîÎìúÎ•º Î¨∂Ïñ¥Ï§ÄÎã§.

df['Ïó∞Î†πÎåÄ'] = df['Ïó∞Î†πÎåÄÏΩîÎìú(5ÏÑ∏Îã®ÏúÑ)'].map(age_code)
df_1 = df.groupby('Ïó∞Î†πÎåÄ')['ÌóàÎ¶¨ÎëòÎ†à'].describe()
df_1
```

# . "ÏùåÏ£ºÏó¨Î∂Ä", "Ìù°Ïó∞ÏÉÅÌÉú", "Ïó∞Î†πÎåÄÏΩîÎìú(5ÏÑ∏Îã®ÏúÑ)", "ÏÑ±Î≥ÑÏΩîÎìú"Ïóê ÎåÄÌïú ÏÉÅÍ¥ÄÍ≥ÑÏàòÎ•º Íµ¨ÌïòÍ≥† ÏãúÍ∞ÅÌôî Ìï¥Ï£ºÏÑ∏Ïöî.
Îã§ÏùåÏùò Í≤∞Í≥ºÍ∞Ä ÎÇòÏò§ÎèÑÎ°ù Ìï©ÎãàÎã§. Ïù¥ Îïå Í∑∏ÎûòÌîÑÏùò ÏÉâÏÉÅ, Ïä§ÌÉÄÏùº ÌëúÌòÑÎ∞©ÏãùÏóê Ï∞®Ïù¥Í∞Ä ÏûàÎçîÎùºÎèÑ ÏÉÅÍ¥ÄÍ≥ÑÏàòÍ∞íÏù¥ Ï∞®Ïù¥Í∞Ä ÏóÜÎã§Î©¥ Î¨∏Ï†ú ÏóÜÏäµÎãàÎã§.


üîΩ Ï∂úÎ†• ÏòàÏãú


```python
df_2 = df[["ÏùåÏ£ºÏó¨Î∂Ä", "Ìù°Ïó∞ÏÉÅÌÉú", "Ïó∞Î†πÎåÄÏΩîÎìú(5ÏÑ∏Îã®ÏúÑ)", "ÏÑ±Î≥ÑÏΩîÎìú"]].corr()
df_2
```

# Ìù°Ïó∞ÌïòÎäî ÏÇ¨ÎûåÍ≥º ÏùåÏ£ºÌïòÎäî ÏÇ¨ÎûåÎì§Ïùò ÏàòÎäî ÏñºÎßàÎÇò Ï∞®Ïù¥Í∞Ä ÏûàÏùÑÍπåÏöî?

Í±¥Í∞ïÎ≥¥ÌóòÏóêÏÑú Ï†úÍ≥µÌïòÍ≥† ÏûàÎäî Îç∞Ïù¥ÌÑ∞Îäî Ïö©ÎüâÏùÑ Ï§ÑÏù¥Í∏∞ ÏúÑÌï¥ ÎåÄÎ∂ÄÎ∂ÑÏùò Îç∞Ïù¥ÌÑ∞Î•º Ïà´Ïûê ÌòïÌÉúÎ°ú Ï†úÍ≥µÌïòÍ≥† ÏûàÏäµÎãàÎã§. <br>
replace Í∏∞Îä•ÏùÑ ÏÇ¨Ïö©ÌïòÎ©¥ ÎîïÏÖîÎÑàÎ¶¨Ïùò Í∞íÏùÑ ÏùòÎØ∏ÌïòÎäî Î¨∏Ïûê ÌòïÌÉúÎ°ú Î≥ÄÍ≤ΩÌï¥ Î≥º Ïàò ÏûàÏäµÎãàÎã§. <br>
Îã§Ïùå Ï†úÍ≥µÎêòÎäî ÎîïÏÖîÎÑàÎ¶¨ Í∞íÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ Ïà´ÏûêÍ∞Ä ÏùòÎØ∏ÌïòÎäî Î¨∏Ïûê ÌòïÌÉúÎ°ú Î≥ÄÍ≤ΩÌïú Îí§ Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞ÏóêÏÑú ÏùåÏ£ºÏôÄ Ìù°Ïó∞Ïóê ÎåÄÌïú ÍµêÏ∞® ÌÖåÏù¥Î∏îÏùÑ Íµ¨Ìï¥ Í∞Å Ìï¥Îãπ ÎπàÎèÑÏàòÎ•º Íµ¨ÌïòÍ≥† ÏãúÍ∞ÅÌôî Ìï¥Ï£ºÏÑ∏Ïöî!

```python
# Ìù°Ïó∞ 1(ÌîºÏö∞ÏßÄ ÏïäÎäîÎã§), 2(Ïù¥Ï†ÑÏóê ÌîºÏõ†ÏúºÎÇò ÎÅäÏóàÎã§), 3(ÌòÑÏû¨ÎèÑ ÌîºÏö∞Í≥† ÏûàÎã§)
# ÏùåÏ£º 0(ÎßàÏãúÏßÄ ÏïäÏùÄÎã§), 1(ÎßàÏã†Îã§)
smoke = {1 : "Ìù°Ïó∞ÏïàÌï®", 2: "ÎÅäÏùå", 3: "Ìù°Ïó∞Ï§ë"}
drink = {0: "ÏïàÎßàÏã¨", 1: "ÎßàÏã¨"}

# df["Ìù°Ïó∞"]
# df["ÏùåÏ£º"]
```

- groupby, pivot_tableÎ°ú Íµ¨Ìï† ÏàòÎèÑ ÏûàÏßÄÎßå, ÏàòÏóÖÏóêÏÑúÎäî Îã§Î£®ÏßÄ ÏïäÏïòÎçò pandasÏùò crosstab Í∏∞Îä•ÏùÑ ÏÇ¨Ïö©Ìï¥ÏÑúÎèÑ ÎπÑÍµêÏ†Å ÏâΩÍ≤å Íµ¨Ìï† Ïàò ÏûàÏäµÎãàÎã§.
- crosstab ÏùÑ ÏÇ¨Ïö©Ìï† ÎïåÎäî Îã§ÏùåÏùò Î¨∏ÏÑúÎ•º Ï∞∏Í≥†ÌïòÏó¨ Íµ¨Ìï¥Î≥¥ÏÑ∏Ïöî. pandas.crosstab ‚Äî pandas documentation
- Í∑∏Î¶¨Í≥† ÎßâÎåÄÍ∑∏ÎûòÌîÑÎ•º ÌÜµÌï¥ Í≤∞Í≥ºÎ•º ÏãúÍ∞ÅÌôîÌï¥Ï£ºÏÑ∏Ïöî! ÏãúÍ∞ÅÌôîÎ•º Ìï† Îïå Í∑∏ÎûòÌîÑÏùò Ïä§ÌÉÄÏùºÏù¥ÎÇò ÎßâÎåÄÏùò ÏÉâÏÉÅ Îì±Ïù¥ Îã¨ÎùºÎèÑ ÏÉÅÍ¥Ä ÏóÜÏúºÎãà ÏÇ¨Ïö©ÌïòÍ≥† Ïã∂ÏùÄ Ïä§ÌÉÄÏùºÏùÑ Ï†ÅÏö©Ìï¥ÏÑú ÏãúÍ∞ÅÌôîÌï¥ÎèÑ Ï¢ãÏäµÎãàÎã§.

üîΩ Ï∂úÎ†• ÏòàÏãú


```python
# Ìù°Ïó∞Í≥º ÏùåÏ£º Ïª¨Îüº ÏÉùÏÑ±

df['Ìù°Ïó∞'] = df['Ìù°Ïó∞ÏÉÅÌÉú'].replace(smoke.keys(), smoke.values())
df['ÏùåÏ£º'] = df['ÏùåÏ£ºÏó¨Î∂Ä'].replace(drink.keys(), drink.values())
```

```python
# crosstabÏúºÎ°ú Îç∞Ïù¥ÌÑ∞ Ï∂îÏ∂úÌïòÍ∏∞
df_3 = pd.crosstab(df['ÏùåÏ£º'], df['Ìù°Ïó∞'])
df_3
```

```python
# Í∑∏ÎûòÌîÑ ÏãúÍ∞ÅÌôî 
sns.countplot(data = df, x = 'Ìù°Ïó∞', hue = 'ÏùåÏ£º')
```

- ÏÑúÎ∏åÌîåÎ°ØÏúºÎ°ú ÏãúÍ∞ÅÌôîÌïòÏó¨ ÌëúÌòÑÌï¥ Ï£ºÍ≥† matplotlib Ïóê ÎåÄÌïú ÏÇ¨Ïö©Î≤ïÏùÄ Îã§ÏùåÏùò ÏòÅÏÉÅÎèÑ Ìï®Íªò Ï∞∏Í≥†Ìï¥ Ï£ºÏÑ∏Ïöî! (https://www.youtube.com/watch?v=ZTRKojTLE8M)

```python
# Ïã§Î™Ö Îç∞Ïù¥ÌÑ∞ ÏÇ≠Ï†ú
not_blind = df[(df['ÏãúÎ†•(Ïö∞)'] != 9.9) & (df['ÏãúÎ†•(Ï¢å)'] != 9.9)]
not_blind
```

```python
sns.barplot(data = not_blind, x='ÏãúÎ†•(Ï¢å)', y = 'Ïó∞Î†πÎåÄ', hue = 'ÏÑ±Î≥ÑÏΩîÎìú')
```
