한스 로슬링(Hans Rosling, 1948년 7월 27일 ~ 2017년 2월 7일)은 스웨덴의 의사이자 통계학자로 비영리 벤처 갭마인더 재단의 공동설립자이기도 합니다. <br>빅데이터를 가장 잘 활용하는 보건 통계학자로 알려져 있습니다. 베스트셀러 책인 "팩트풀니스" 저자이기도 합니다. <br>갭마인더 사이트에서는 연도별, 국가별 GDP와 기대수명 데이터를 제공하고 있는데, 대표적으로 파이썬 라이브러리 중 'seaborn'에서 제공되는 예제 데이터가 있습니다. <br>오늘은 이 데이터를 활용해 문제를 풀어보겠습니다.

**다음의 데이터는 연도, 국가별 기대수명을 나타내고 있는 데이터 입니다. 2011년 부터의 연도별, 국가별 평균 기대수명을 구해주세요. (2011년도 포함되게 구합니다.)**

[필수 조건] 
- groupby 나 pivot_table을 활용합니다. groupby 로 구할 때 unstack()이라는 기능을 사용하여 아래와 같이 컬럼에 인덱스 값을 올려서 표기할 수 있습니다.
- 만약 행에는 '연도'가, 열에 '국가'가 들어있고 수치 데이터의 결과값이 아래와 같다면, 출력형태는 조금 달라도 괜찮습니다.
- 데이터는 다음의 방법으로 읽어옵니다.


```python
import pandas as pd
life_df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")
life_df
```

잘 모르겠다면, 다음의 판다스 문서를 참고해서 풀어보세요!<br> => https://pandas.pydata.org/docs/getting_started/intro_tutorials/07_reshape_table_layout.html

```python
life_df.info()
```

```python
# 국가 개수 파악
# 연도마다 국가의 개수가 다름을 확인
life_df['Country'].value_counts()
```

```python
# 연도별 조사 국가 개수 파악
# 2011년도 이후부터는 6개로 동일하기 때문에 2011년 이후의 데이터만 살리면 전처리하기 쉬울 것으로 판단됨
life_df.groupby(['Year'])['Country'].count()
```

```python
# 2011년도 이전 데이터 삭제
life_df.drop(life_df[(life_df['Year'] < 2011)].index, inplace=True)
life_df
```

```python
# pivot_table 함수로 index는 연도, columns는 국가, 값은 평균 기대수명
life_df.pivot_table(index='Year', columns='Country', values = 'Life_Expectancy')
```

**데이터 형태**
1. Boolean 
    - 파이썬에서는 명시적인 것이 암시적인 것보다 낫다. 
    - True나 False는 0과 1로도 명시 가능

2. Number & String
    - 숫자 1과 문자 "1"은 다름
    - type() 명령어를 통해 데이터 타입 확인 가능
<br><br>
3. Strings and list
    - lower 함수로 문자열 전부를 소문자로 변환 가능
    - upper 함수로 문자열 전부를 대문자로 변환 가능
    - list에는 append 함수로 원소 추가 가능
    - list는 indexing이 가능하고, 번호는 0부터 시작
    - 마지막 인덱스는 -1로 표현

4. Control Flow
    - 제어문, 조건문, 반복문
    - 문자열의 공백을 제거하거나, 공백으로 문자열을 분리하여 리스트에 담을 수 있음
    - len 함수로 문자열의 길이 혹은 리스트 원소의 개수를 알 수 있음
    - slicing으로 문자 가져오기 가능
    - startwith 함수로 특정 문자로 시작하는지 알 수 있음
    - in 함수로 특정 문자의 포함 여부 확인 가능
    - join 함수로 리스트 내의 모든 문자열 합치기 가능
    

**Pandas**
1. Dataframe
    - 2차원 구조, 행렬
    - pd.DataFrame? 를 실행하면 도움말 나옴
<br><br>
2. Series
    - 1차원 구조, 벡터
<br><br>
3. Subset
    - 두 개 이상의 값을 불러올 때 Series형태로 불러올 경우 키값 오류 발생
    - Columns 이름을 리스트 형태로 작성
        - df[['a', 'b', 'c']]
<br><br>
4. Summerize Data
    - Categorical한 값의 빈도수를 구하는 방법
        - df["a"].value_counts()
<br><br>
5. Reshaping
    - "a"컬럼을 기준으로 정렬하기
        - df["a"].sort_values()

        - df.sort_values("a")
    - 역순으로 정렬하기

    - c 컬럼 삭제

<br><br>
6. Group data
    - a" 컬럼값을 Groupby하여 "b"의 컬럼값 평균값 구하기 
        - df.groupby(["a"])["b"].mean()

        - pd.pivot_table(df, index="a", values = "b", aggfunc = "sum")
