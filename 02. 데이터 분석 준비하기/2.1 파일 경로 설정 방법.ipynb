# 여러 파일 한 번에 로드하기

```python
# store 폴더의 csv 파일만 불러오기 
file_csv = glob("data/store/*.csv")
file_csv
```

```python
# '소상'으로 시작하는 파일만 로드하기
# 운영체제마다 한글 인코딩이 달라서 직접 타이핑 했을 때 원하는 결과가 나오지 않으면 드래그해서 복붙할 것 
glob("data/store/소상*.csv")
```

```python
glob("data/store/*_*_*.csv")
```

```python
# 여러 파일 이름 가져오기 
file_list = []

for file_csv_name in file_csv:
    # 인코딩 필요 여부 확인 후 작성하기
    temp = pd.read_csv(file_csv_name)
    file_list.append(temp)
```

```python
# 마지막 데이터 프레임까지 불러오기
file_list[-1]
```

```python
len(file_list)
```

```python
# 컬럼명이 같으면 하나의 데이터프레임에 합쳐줌
df = pd.concat(file_list)
df.shape
```
